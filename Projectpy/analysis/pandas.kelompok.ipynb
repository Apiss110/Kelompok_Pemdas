```python
#no 17
kolom_kategoridm = {}
def kategori(x):
    if x < 50000: return "Rendah"
    elif x < 100000: return "Sedang"
    return "Tinggi"

df["kategori_dm"] = df["jumlah_penderita_dm"].apply(kategori)
df.head()
```

```python
#no 18
kolom_persentase_tahun = df['persentase_tahun'] = df['jumlah_penderita_dm'] / df.groupby('tahun')['jumlah_penderita_dm'].transform('sum') * 100
df.head()
```

```python
#no 19
tabel_ringkas = df.groupby("tahun").agg(
 total_jumlah=('jumlah_penderita_dm','sum'),
 jumlah_kabupaten=('nama_kabupaten_kota','count')
)
tabel_ringkas
```

```python
#D
df_2019 = df[df['tahun'] == 2019]

plt.figure(figsize=(12,6))
plt.bar(df_2019['nama_kabupaten_kota'], df_2019['jumlah_penderita_dm'])
plt.xticks(rotation=90)
plt.title('Jumlah Penderita DM per Kabupaten/Kota Tahun 2019')
plt.xlabel('Kabupaten/Kota')
plt.ylabel('Jumlah Penderita DM')
plt.tight_layout()
plt.show()

```

```python
total_per_year = df.groupby('tahun')['jumlah_penderita_dm'].sum().reset_index()

plt.figure(figsize=(7,4))
plt.plot(total_per_year['tahun'], total_per_year['jumlah_penderita_dm'], marker='o', linewidth=2)
plt.title('Total Jumlah Penderita DM Jawa Barat per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Jumlah DM')
plt.grid(True)
plt.show()

total_per_year
```
