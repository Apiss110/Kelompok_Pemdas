```python
#nomer 3
data2 = df[df['tahun']==2021]
data2 = data2[['nama_kabupaten_kota','jumlah_sampah','tahun']]
data2
```

```python
sampah_tinggi = data2.nlargest(5, ['jumlah_sampah'])
sampah_tinggi
```

```python
#nomer 4
data3 = df.groupby('nama_kabupaten_kota')['jumlah_sampah']
data3_df = pd.DataFrame({
    'nama_kabupaten_kota': data3.mean().index,
    'mean_jumlah_sampah': data3.mean().values,
    'median_jumlah_sampah': data3.median().values
}).sort_values('mean_jumlah_sampah', ascending=False)
data3_df
```

```python
top3_mean = data3_df.head(3)
top3_mean
```

```python
#nomer 5
total_2016 = df[df['tahun']==2016][['jumlah_sampah']].sum()
total_2016
```

```python
total_2023 = df[df['tahun']==2023][['jumlah_sampah']].sum()
total_2023
```

```python
selisih = total_2016 - total_2023
selisih
```

```python
persen = (selisih/total_2016)*100
persen
```

```python
#nomer 6
data_2021 = df[df['tahun']==2021].copy()
total_2021 = data_2021[['jumlah_sampah']].sum()
total_2021
```

```python
data_2021[["persentase_kontribusi"]] = (data_2021[['jumlah_sampah']] / total_2021) * 100
data_2021
```

```python
top10_2021 = data_2021.sort_values("persentase_kontribusi", ascending=False).head(10)
top10_2021
```
